```python
import pandas as pd
from sodapy import Socrata
import os
from dotenv import load_dotenv

load_dotenv('.env')

TOKEN_APP = os.getenv("TOKEN_SODAPY")

client = Socrata("www.datos.gov.co", 
                 TOKEN_APP)
```

```python
client.timeout = 10000
```

```python
socrata_data_identifier = "jbjy-vk9h"

Query = """ 
select
    id_contrato, nombre_entidad, departamento, descripcion_del_proceso, valor_del_contrato
where
    fecha_de_firma > '2024-01-01'
limit
10000000
"""

result = client.get(socrata_data_identifier, query = Query)

result_df = pd.DataFrame.from_records(result)
```
